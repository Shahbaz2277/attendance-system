```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.title("Two Lines")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

hours = [2, 3, 4, 5, 6, 7, 8]
grades = [60, 65, 70, 75, 80, 85, 90]

plt.scatter(hours, grades, color='blue')
plt.title("Study Hours vs Grades")
plt.xlabel("Study Hours")
plt.ylabel("Grade")
plt.show()
```

```python
import matplotlib.pyplot as plt

hours = [2, 3, 4, 5, 6, 7, 8]
grades = [60, 65, 70, 75, 80, 85, 90]

plt.scatter(hours, grades, color='red', s=100, marker='o')
plt.title("Study Hours vs Grades")
plt.xlabel("Study Hours")
plt.ylabel("Grade")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

products = ['A', 'B', 'C', 'D', 'E']
sales = [100, 150, 80, 120, 90]

plt.bar(products, sales, color='green')
plt.title("Product Sales")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

```python
import matplotlib.pyplot as plt

products = ['A', 'B', 'C', 'D', 'E']
sales = [100, 150, 80, 120, 90]

plt.barh(products, sales, color='orange')
plt.title("Product Sales")
plt.xlabel("Sales")
plt.ylabel("Product")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
grades = np.random.normal(70, 10, 100)

plt.hist(grades, bins=10, color='purple', edgecolor='black')
plt.title("Grade Distribution")
plt.xlabel("Grade")
plt.ylabel("Number of Students")
plt.show()
```

```python
import matplotlib.pyplot as plt

companies = ['Company A', 'Company B', 'Company C', 'Others']
shares = [40, 30, 20, 10]

plt.pie(shares, labels=companies, autopct='%1.1f%%')
plt.title("Market Share")
plt.show()
```

```python
import matplotlib.pyplot as plt

companies = ['Company A', 'Company B', 'Company C', 'Others']
shares = [40, 30, 20, 10]
explode = [0.1, 0, 0, 0]
plt.pie(shares, labels=companies, autopct='%1.1f%%', explode=explode)
plt.title("Market Share")
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot([1, 2, 3, 4], [1, 4, 9, 16])
axes[0, 0].set_title("Line Plot")

axes[0, 1].scatter([1, 2, 3, 4], [1, 4, 9, 16])
axes[0, 1].set_title("Scatter Plot")

axes[1, 0].bar(['A', 'B', 'C'], [10, 20, 15])
axes[1, 0].set_title("Bar Chart")

axes[1, 1].pie([30, 40, 30], labels=['X', 'Y', 'Z'])
axes[1, 1].set_title("Pie Chart")

plt.tight_layout()
plt.show()
```
